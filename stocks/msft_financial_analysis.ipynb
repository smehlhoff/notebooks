# Import Python Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import Financial Data for MSFT

```python
# https://www.wsj.com/market-data/quotes/MSFT/financials

income = pd.read_excel("./data/msft.xlsx", sheet_name="income")
balance = pd.read_excel("./data/msft.xlsx", sheet_name="balance")
cash = pd.read_excel("./data/msft.xlsx", sheet_name="cash")
```

```python
balance.set_index("Year")
```

# Income Statement Categories

## Calculate Revenue & Growth

```python
revenue_growth = {}

for i in range(0, 4):
    revenue_growth.update({income.iloc[i]["Year"]: 
                           income.iloc[i]["Sales/Revenue"] / income.iloc[i + 1]["Sales/Revenue"] - 1})
```

```python
revenue_growth
```

```python
x = [x for x in revenue_growth.keys()]
y = [x for x in revenue_growth.values()]

plt.figure(figsize=(5, 3), dpi=150)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.plot(x, y, "bo-", label="MSFT")
plt.plot(x, p(x), ls="--", color="red")

plt.title("Revenue Growth %")
plt.xlabel("Year")
plt.ylabel("Revenue %")

plt.locator_params(integer=True)

plt.legend(loc=0)

plt.show()
```

## Calculate Inventory Turnover

```python
inventory_turnover = {}

for i in range(0, 4):
    inventory_turnover.update({
        income.iloc[i]["Year"]: income.iloc[i]["Cost of Goods Sold (COGS) incl. D&A"] / np.average([
            balance.iloc[i]["Inventories"], 
            balance.iloc[i + 1]["Inventories"]
        ])
    })
```

```python
inventory_turnover
```

```python
x = [int(x) for x in inventory_turnover.keys()]
y = [int(x) for x in inventory_turnover.values()]

plt.figure(figsize=(5, 3), dpi=150)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.plot(x, y, "bo-", label="MSFT")
plt.plot(x, p(x), ls="--", color="red")

plt.title("Inventory Turnover")
plt.xlabel("Year")
plt.ylabel("Turnover Ratio")

plt.locator_params(integer=True)

plt.legend(loc=0)

plt.show()
```

## Calculate Inventory-to-Revenue

```python
inventory_to_revenue = {}

for i in range(0, 4):
    inventory_to_revenue.update({
        income.iloc[i]["Year"]: np.average([
            balance.iloc[i]["Inventories"], 
            balance.iloc[i + 1]["Inventories"]
        ]) / income.iloc[i]["Sales/Revenue"]
    })
```

```python
inventory_to_revenue
```

```python
x = [int(x) for x in inventory_to_revenue.keys()]
y = [float(x) for x in inventory_to_revenue.values()]

plt.figure(figsize=(5, 3), dpi=150)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.plot(x, y, "bo-", label="MSFT")
plt.plot(x, p(x), ls="--", color="red")

plt.title("Inventory-to-Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue Ratio")

plt.locator_params(integer=True)

plt.legend(loc=0)

plt.show()
```

## Calculate Days Inventory

```python
days_inventory = {}

for i in range(0, 4):
    days_inventory.update({
        income.iloc[i]["Year"]: np.average([
            balance.iloc[i]["Inventories"], 
            balance.iloc[i + 1]["Inventories"]
        ]) / income.iloc[i]["Cost of Goods Sold (COGS) incl. D&A"] * 365
    })
```

## Save Inventory Ratios to DataFrame

## DuPont Analysis (3-step method)

DuPont Analysis = Net Profit Margin x AT x EM

That is,

- Net Profit Margin = Net Income / Revenue
- AT = Asset Turnover = Revenue / Average Total Assets
- EM = Equity Multiplier (or financial leverage) = Average Total Assets / Average Shareholders' Equity

Source: https://www.investopedia.com/terms/d/dupontanalysis.asp

## Calculate Net Margin

```python
net_margins = {}

for i in range(0, 4):
    net_margins.update({income.iloc[i]["Year"]: income.iloc[i]["Net Income"] / income.iloc[i]["Sales/Revenue"]})
```

```python
net_margins
```

```python
x = [int(x) for x in net_margins.keys()]
y = [float(x) for x in net_margins.values()]

plt.figure(figsize=(5, 3), dpi=150)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.plot(x, y, "bo-", label="MSFT")
plt.plot(x, p(x), ls="--", color="red")

plt.title("Net Margin")
plt.xlabel("Year")
plt.ylabel("Net Margin %")

plt.locator_params(integer=True)

plt.legend(loc=0)

plt.show()
```

## Calculate Asset Turnover

```python
asset_turnover = {}

for i in range(0, 4):
    asset_turnover.update({income.iloc[i]["Year"]: income.iloc[i]["Sales/Revenue"] / np.average([
        balance.iloc[i]["Total Assets"], 
        balance.iloc[i + 1]["Total Assets"]
    ])})
```

```python
asset_turnover
```

```python
x = [int(x) for x in asset_turnover.keys()]
y = [float(x) for x in asset_turnover.values()]

plt.figure(figsize=(5, 3), dpi=150)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.plot(x, y, "bo-", label="MSFT")
plt.plot(x, p(x), ls="--", color="red")

plt.title("Asset Turnover")
plt.xlabel("Year")
plt.ylabel("Asset Turnover %")

plt.locator_params(integer=True)

plt.legend(loc=0)

plt.show()
```

## Calculate Equity Multiplier

```python
equity_multiplier = {}

for i in range(0, 4):
    equity_multiplier.update({income.iloc[i]["Year"]: 
                              np.average([
                                  balance.iloc[i]["Total Assets"], 
                                  balance.iloc[i + 1]["Total Assets"]
                              ]) / np.average([
                                  balance.iloc[i]["Total Shareholders' Equity"], 
                                  balance.iloc[i + 1]["Total Shareholders' Equity"]
                              ])})
```

```python
equity_multiplier
```

## Calculate DuPont

```python
dupont = {}

for year in equity_multiplier.keys():
    dupont.update({year: net_margins[year] * asset_turnover[year] * equity_multiplier[year]})
```

```python
dupont
```

```python
x = [x for x in dupont.keys()]
y = [x for x in dupont.values()]

plt.figure(figsize=(5, 3), dpi=150)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.plot(x, y, "bo-", label="MSFT")
plt.plot(x, p(x), ls="--", color="red")

plt.title("DuPont Analysis")
plt.xlabel("Year")
plt.ylabel("DuPont %")

plt.locator_params(integer=True)

plt.legend(loc=0)

plt.show()
```

## Save DuPont Ratios to DataFrame

```python
ratios["Net Margin %"] = pd.Series([x for x in net_margins.values()])
ratios["Asset Turnover"] = pd.Series([x for x in asset_turnover.values()])
ratios["Equity Multiplier"] = pd.Series([x for x in equity_multiplier.values()])
ratios["DuPont Analysis"] = pd.Series([x for x in dupont.values()])

ratios.set_index("Year")
```
