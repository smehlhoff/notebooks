# DuPont Analysis

DuPont Analysis = Net Profit Margin * AT * EM

That is,

- Net Profit Margin = Net Income / Revenue
- AT = Asset Turnover = Revenue / Average Total Assets
- EM = Equity Multiplier (or financial leverage) = Average Total Assets / Average Shareholders' Equity

Source: https://www.investopedia.com/terms/d/dupontanalysis.asp

## Calculate Net Margin for DuPont Analysis

```python
aapl_net_margins = {}

for i in range(0, 4):
    aapl_net_margins.update({aapl_income.iloc[i]["Year"]: 
                        aapl_income.iloc[i]["Net Income"] 
                        / aapl_income.iloc[i]["Sales/Revenue"]})
    
msft_net_margins = {}

for i in range(0, 4):
    msft_net_margins.update({msft_income.iloc[i]["Year"]: 
                        msft_income.iloc[i]["Net Income"] 
                        / msft_income.iloc[i]["Sales/Revenue"]})
```

```python
aapl_net_margins
```

```python
msft_net_margins
```

```python
x = [x for x in aapl_net_margins.keys()]
y1 = [x for x in aapl_net_margins.values()]
y2 = [x for x in msft_net_margins.values()]

plt.figure(figsize=(5, 3), dpi=150)

plt.plot(x, y1, "ro-", label="AAPL")
plt.plot(x, y2, "bo-", label="MSFT")

plt.title("Net Margin %")
plt.xlabel("Year")
plt.ylabel("Percent")

plt.locator_params(integer=True)

plt.legend(loc=0)

plt.show()
```

## Calculate Asset Turnover for DuPont Analysis

```python
aapl_asset_turnover = {}

for i in range(0, 4):
    aapl_asset_turnover.update({aapl_income.iloc[i]["Year"]: aapl_income.iloc[i]["Sales/Revenue"] / np.average([
        aapl_balance.iloc[i]["Total Assets"], 
        aapl_balance.iloc[i + 1]["Total Assets"]
    ])})
    
msft_asset_turnover = {}

for i in range(0, 4):
    msft_asset_turnover.update({msft_income.iloc[i]["Year"]: msft_income.iloc[i]["Sales/Revenue"] / np.average([
        msft_balance.iloc[i]["Total Assets"], 
        msft_balance.iloc[i + 1]["Total Assets"]
    ])})
```

```python
aapl_asset_turnover
```

```python
msft_asset_turnover
```

```python
x = [x for x in aapl_asset_turnover.keys()]
y1 = [x for x in aapl_asset_turnover.values()]
y2 = [x for x in msft_asset_turnover.values()]

plt.figure(figsize=(5, 3), dpi=150)

plt.plot(x, y1, "ro-", label="AAPL")
plt.plot(x, y2, "bo-", label="MSFT")

plt.title("Asset Turnover %")
plt.xlabel("Year")
plt.ylabel("Percent")

plt.locator_params(integer=True)

plt.legend(loc=0)

plt.show()
```

## Calculate Equity Multiplier for DuPont Analysis

```python
aapl_equity_multiplier = {}

for i in range(0, 4):
    aapl_equity_multiplier.update({aapl_income.iloc[i]["Year"]: 
                              np.average([
                                  aapl_balance.iloc[i]["Total Assets"], 
                                  aapl_balance.iloc[i + 1]["Total Assets"]
                              ]) / np.average([
                                  aapl_balance.iloc[i]["Total Shareholders' Equity"], 
                                  aapl_balance.iloc[i + 1]["Total Shareholders' Equity"]
                              ])})
```

```python
msft_equity_multiplier = {}

for i in range(0, 4):
    msft_equity_multiplier.update({msft_income.iloc[i]["Year"]: 
                              np.average([
                                  msft_balance.iloc[i]["Total Assets"], 
                                  msft_balance.iloc[i + 1]["Total Assets"]
                              ]) / np.average([
                                  msft_balance.iloc[i]["Total Shareholders' Equity"], 
                                  msft_balance.iloc[i + 1]["Total Shareholders' Equity"]
                              ])})
```

```python
aapl_equity_multiplier
```

```python
msft_equity_multiplier
```

## Calculate DuPont

```python
aapl_dupont = {}

for year in aapl_equity_multiplier.keys():
    aapl_dupont.update({year: aapl_net_margins[year] * aapl_asset_turnover[year] * aapl_equity_multiplier[year]})
    
msft_dupont = {}

for year in aapl_equity_multiplier.keys():
    msft_dupont.update({year: msft_net_margins[year] * msft_asset_turnover[year] * msft_equity_multiplier[year]})
```

```python
aapl_dupont
```

```python
msft_dupont
```

```python
x = [x for x in aapl_dupont.keys()]
y1 = [x for x in aapl_dupont.values()]
y2 = [x for x in msft_dupont.values()]

plt.figure(figsize=(5, 3), dpi=150)

plt.plot(x, y1, "ro-", label="AAPL")
plt.plot(x, y2, "bo-", label="MSFT")

plt.title("DuPont Analysis")
plt.xlabel("Year")
plt.ylabel("Percent")

plt.locator_params(integer=True)

plt.legend(loc=0)

plt.show()
```
