```python
import pandas as pd
import numpy as np
```

```python
from numpy.random import randn

np.random.seed(101)
rand_mat = randn(5, 4)
```

```python
rand_mat
```

```python
df = pd.DataFrame(data=rand_mat, index="A B C D E".split(), columns="W X Y Z".split())
```

```python
# this is more conventional

df[["W", "Y"]]
```

```python
df["NEW"] = df["W"] + df["Y"]
```

```python
df
```

```python
# setting inplace = True will make this permanent

df.drop("NEW", axis=1, inplace=True)
```

```python
df["W"] > 0
```

```python
# return > 0 for W column; return all other rows

df[df["W"] > 0]
```

```python
cond_1 = df["W"] > 0
cond_2 = df["Y"] > 1
```

```python
# the AND or OR operators are not designed for pandas series

df[cond_1 and cond_2]
```

```python
# the AND operator for pandas series

df[(cond_1) & (cond_2)]
```

```python
# the OR operator for pandas series

df[(cond_1) | (cond_2)]
```

```python
# combined into one line

df[(df["W"] > 0) & (df["Y"] > 1)]
```
