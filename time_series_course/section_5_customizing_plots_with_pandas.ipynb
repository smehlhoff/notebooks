```python
import pandas as pd
```

```python
df2 = pd.read_csv("./data/section_5/data2.csv")
```

```python
df2["c"].plot.line(figsize=(10, 5), ls="--", c="red", lw=3)
```

```python
title = "My Line Chart"
x_label = "X Label"
y_label = "Y Label"

ax = df2["c"].plot.line(title=title, figsize=(10, 5))

ax.set(xlabel=x_label, ylabel=y_label)
```

```python
# legend location
# 0 - best
# 1 - upper right
# 2 - upper left
# 3 - lower left
# 4 - lower right
# 5 - right
# 6 - center left
# 7 - center right
# 8 - lower center
# 9 - upper center
# 10 - center

ax = df2.plot()

ax.legend(loc=4)
```
