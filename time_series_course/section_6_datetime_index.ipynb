```python
import numpy as np
```

```python
# the datetime64[D] represents day level precision
# so, if you needed a year level precision, set dtype="datetime64[Y]"

np.array(["2020-03-15", "2020-03-16", "2020-03-17"], dtype="datetime64")
```

```python
# a datetime range between 2018-06-01 and 2018-06-23, with a step size of 7 days

np.arange("2018-06-01", "2018-06-23", 7, dtype="datetime64[D]")
```

```python
np.arange('2000', "2018", dtype="datetime64[Y]")
```

```python
import pandas as pd
```

```python
# DatetimeIndex with nanosec precision by default

pd.date_range("2020-01-01", periods=7, freq="D")
```

```python
# pandas can infer the date parameter

pd.date_range("Jan 01, 2018", periods=7, freq="D")
```

```python
pd.to_datetime(["1/2/2018", "Jan 03, 2018"])
```

```python
# changing international dates to U.S.

pd.to_datetime(["2/1/2018", "3/1/2018"], format="%d/%m/%Y")
```

```python
data = np.random.randn(3, 2)
```

```python
data
```

```python
index = pd.date_range("2020-01-01", periods=3, freq="D")
columns = ["A", "B"]

df = pd.DataFrame(data, index=index, columns=cols)
```

```python
df
```

```python
# pandas knows the index is a datetime, not string attribute

df.index
```

```python
# return the most recent date

df.index.max()
```

```python
# return the index position for the most recent date

df.index.argmax()
```

```python
# return the oldest date

df.index.min()
```

```python
# return the index position for the oldest date

df.index.argmin()
```
