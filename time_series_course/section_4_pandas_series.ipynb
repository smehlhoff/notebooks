```python
# numpy vs pandas

# numpy uses arrays with an index (e.g., 0, 1, 2)
# pandas series is built on top of numpy arrays; therefore, it can use arrays with an index
# however, it can also use named columns (think database tables)
```

```python
import numpy as np
import pandas as pd
```

```python
labels = ["a", "b", "c"]
```

```python
my_list = [10, 20, 30]
```

```python
arr = np.array(my_list)
```

```python
d = {"a": 10, "b": 20, "c": 30}
```

```python
pd.Series(data=my_list)
```

```python
pd.Series(data=arr, index=labels)
```

```python
pd.Series(data=[10, "a", 4.4])
```

```python
s1 = pd.Series(data=[1, 2, 3, 4], index=["USA", "Germany", "USSR", "Japan"])
```

```python
s1
```

```python
s1["USA"]
```

```python
s2 = pd.Series(data=[1, 4, 5, 6], index=["USA", "Germany", "Italy", "Japan"])
```

```python
s2
```

```python
s1 + s2
```
