```python
arr_2d
```

```python
arr_2d[1][1]
```

```python
# return rows between 0 and 2 (not including 2)

arr_2d[:2]
```

```python
# return rows between 0 and 2 (not including 2); start at column 1 and select everything thereafter

arr_2d[:2, 1:]
```

```python
# conditional selection

arr = np.arange(1, 11)
```

```python
arr
```

```python
arr > 4
```

```python
bool_array = arr > 4
```

```python
bool_array
```

```python
# returns values where bool is True

arr[bool_array]
```

```python
# steps above combined into one line

arr[arr > 4]
```

```python
arr[arr == 2]
```

```python
arr[arr <= 6]
```
