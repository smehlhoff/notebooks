```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("./data/gas_prices.csv")
```

```python
data
```

```python
x = data["Year"]
y1 = data["USA"]
y2 = data["Canada"]
y3 = data["Germany"]

plt.figure(figsize=(5, 3), dpi=150)

plt.plot(x, y1, label="USA", ls="-", color="blue", )
plt.plot(x, y2, label="Canada", ls="-", color="red")
plt.plot(x, y3, label="Germany", ls="-", color="green")

plt.title("Gas Prices in USD", fontdict={"fontname": "Times New Roman", "fontsize": 24})
plt.xlabel("Year", fontdict={"fontname": "Times New Roman", "fontsize": 16})
plt.ylabel("Price", fontdict={"fontname": "Times New Roman", "fontsize": 16})

plt.locator_params(integer=True)

plt.xticks(data["Year"][::3])

plt.legend(loc=0)

plt.show()
```

```python
data["USA"].describe()
```

```python
x = data["Year"]
y1 = data["USA"]

z = np.polyfit(x, y1, 1)
p = np.poly1d(z)

plt.figure(figsize=(5, 3), dpi=150)

plt.plot(x, y1, label="USA", ls="-", color="blue", marker="o")
plt.plot(x, p(x), ls="--", color="red")

plt.title("Gas Prices in USD", fontdict={"fontname": "Times New Roman", "fontsize": 24})
plt.xlabel("Year", fontdict={"fontname": "Times New Roman", "fontsize": 16})
plt.ylabel("Price", fontdict={"fontname": "Times New Roman", "fontsize": 16})

plt.locator_params(integer=True)

plt.xticks(data["Year"][::3].tolist() + [2011])

plt.legend(loc=0)

plt.show()
```

```python
plt.hist(data["USA"], edgecolor="k", bins=5)

plt.title("USA Gas Price", fontdict={"fontname": "Times New Roman", "fontsize": 24})
plt.xlabel("Price", fontdict={"fontname": "Times New Roman", "fontsize": 16})
plt.ylabel("Frequency", fontdict={"fontname": "Times New Roman", "fontsize": 16})

plt.show()
```

```python
fifa = pd.read_csv("./data/fifa_data.csv")
```

```python
fifa.head(5)
```

```python
barcelona = fifa.loc[fifa["Club"] == "FC Barcelona"]
juventus = fifa.loc[fifa["Club"] == "Juventus"]
manchester_united = fifa.loc[fifa["Club"] == "Manchester United"]
manchester_city = fifa.loc[fifa["Club"] == "Manchester City"]
san_jose_earthquakes = fifa.loc[fifa["Club"] == "San Jose Earthquakes"]
```

```python
teams = [
    barcelona["Overall"], 
    juventus["Overall"], 
    manchester_united["Overall"], 
    manchester_city["Overall"],
    san_jose_earthquakes["Overall"]
]

plt.figure(figsize=(5, 3), dpi=150)

labels = ["Barcelona", "Juventus", "Manchester United", "Manchester City", "San Jose Earthquakes"]

plt.boxplot(teams, labels=labels)

plt.title("Team Overall Score", fontdict={"fontname": "Times New Roman", "fontsize": 24})
plt.ylabel("Score", fontdict={"fontname": "Times New Roman", "fontsize": 16})

plt.xticks(rotation=45)

plt.show()
```
