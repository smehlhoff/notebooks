```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Line Graphs

```python
x1 = [1, 2, 3, 4]
x2 = [1.5, 2.5, 3.5, 4.5]
y = [2, 4, 6, 8]

plt.plot(x1, y, label="x1 line")
plt.plot(x2, y, label="x2 line", ls="--")

plt.title("My Line Chart", fontdict={"fontname": "Times New Roman", "fontsize": 24})
plt.xlabel("My X Axis", fontdict={"fontname": "Times New Roman", "fontsize": 12})
plt.ylabel("My Y Axis", fontdict={"fontname": "Times New Roman", "fontsize": 12})

plt.legend(loc=4)

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

plt.show()
```

```python
# shorthand notation
# fmt = "[color][marker][line]"

x1 = [1, 2, 3, 4]
x2 = [1.5, 2.5, 3.5, 4.5]
y = [2, 4, 6, 8]

plt.plot(x1, y, "rD:", label="x1 line")
plt.plot(x2, y, "rD:", label="x2 line")

plt.title("My Line Chart", fontdict={"fontname": "Times New Roman", "fontsize": 24})
plt.xlabel("My X Axis", fontdict={"fontname": "Times New Roman", "fontsize": 12})
plt.ylabel("My Y Axis", fontdict={"fontname": "Times New Roman", "fontsize": 12})

plt.legend(loc=4)

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

# set dpi at 300 for high resolution
plt.savefig("my_line_graph.png", dpi=300)

plt.show()
```

```python
x1 = [1, 2, 3, 4]
x2 = np.arange(0, 4.5, 0.5)
y = [2, 4, 6, 8]

plt.figure(figsize=(5, 3), dpi=150)

plt.plot(x1, y, color="red", label="2x")
plt.plot(x2, x2**2, "b^--", label="x^2")

plt.title("My Line Chart", fontdict={"fontname": "Times New Roman", "fontsize": 24})
plt.xlabel("My X Axis", fontdict={"fontname": "Times New Roman", "fontsize": 12})
plt.ylabel("My Y Axis", fontdict={"fontname": "Times New Roman", "fontsize": 12})

plt.legend(loc=4)

plt.xticks([0, 1, 2, 3, 4])

plt.show()
```

# Bar Charts

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ["/", "O", "*"]

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```
